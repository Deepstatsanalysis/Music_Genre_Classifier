# Feature Extraction
from the downloaded music samples using Librosa library

```python
data_dir = './genres/'
genres=['blues','classical','country','disco','hiphop','jazz','metal','pop','reggae','rock']
```

```python
wav_files=glob(data_dir+genres[0]+'/*.wav')
for i in range(1,len(genres),1):
    wav_files= wav_files+glob(data_dir+genres[i]+'/*.wav')
print(len(wav_files))
```

# Extraction using Libraries
and extracted to FinalDataset.csv file

```python
with open('FinalDataset.csv','w',newline='') as f:
    fieldnames = ['index','name','tempo','zcr','spec_bandwidth','rms','spec_centroid','spec_rolloff','chrom_spec']+ ['mfcc' + str(i) for i in range(1,51)]
    
    thewriter=csv.writer(f)
    thewriter.writerow(fieldnames)
    
    print('start')
    for i in range(0,len(wav_files),1):
        x, sr = librosa.load(wav_files[i])
                
        spectral_centroids = librosa.feature.spectral_centroid(x, sr=sr)
        rolloff = librosa.feature.spectral_rolloff(x, sr=sr)
        spec_bw = librosa.feature.spectral_bandwidth(x, sr)
        rms = librosa.feature.rms(x)
        
        
        chromagram = librosa.feature.chroma_stft(x, sr=sr)
        
        tem=librosa.beat.tempo(x,sr)
        
        zcr = librosa.feature.zero_crossing_rate(x)
        
        mfcc = librosa.feature.mfcc(x, sr, n_mfcc=50)
        mfcc_list=[]
        for f in mfcc :
            mfcc_list += [np.mean(f)]
        thewriter.writerow([i,wav_files[i].split('/')[2].split("\\")[0],np.mean(chromagram),np.mean(spectral_centroids),np.mean(spec_bw),np.mean(rolloff),np.mean(zcr),np.median(tem),np.mean(rms),]+mfcc_list)
        print(i)
    print('complete')
```
